```python
import numpy as np  
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("Data1/Titanic-Dataset.csv", index_col = "PassengerId")
df.head(20)
```

```python
df.info()
df.describe()
```

```python
duplicados = df.duplicated().sum()
if duplicados > 0:
    print(f'Se encontraron {duplicados} filas duplicadas. Procediendo a eliminarlas')
    df.drop_duplicates(inplace=True)
else:
    print('No se encontraron filas duplicadas. No es necesario eliminarlas')

print('Dimensiones tras el paso de duplicados')
```

```python
nulos_columnas = df.isnull().sum()
nulos_filas = df[df.isnull().any(axis=1)]
total_nulos = nulos_columnas.sum()
```
