```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/All_Pokemon.csv", index_col = "Number")
```

 Información general del Dataset

```python
df.info()
```

Información estadística 

```python
df.describe()
```

Cantidad de filas duplicadas 

```python
df_duplicados = df.duplicated()
cantidad_duplicados = df_duplicados.sum()
print(f'La cantidad de duplicados es de {cantidad_duplicados}')
```

Dataframe sin filas duplicadas

```python
df_sin_duplicados = df.drop_duplicates()
df_sin_duplicados.head()
```

Contar todos los valores nulos existentes en el dataset

```python
df.isnull()
```

Eliminando valores nulos

```python
df_sin_nulos = df.dropna()
df_sin_nulos
```

Función que permita filtrar los pokemons de tipo fuego

```python
def filtro_Fire(Type):
    return 'Fire' in Type

datos_fuego = df_sin_nulos[df_sin_nulos['Type 1'].apply(filtro_Fire)]

Tipo_Fuego = datos_fuego[['Name','Type 1']]
Tipo_Fuego.head(20)
```

Promedio de pokemons tipo fuego

```python
Promedio_tipo_fuego = datos_fuego [['Def']]
prom_fuego = Promedio_tipo_fuego.mean()
print(f'El promedio de defensa de los pokemons de tipo fuego es de {prom_fuego}')
```

Grafico pokemons tipo fuego

```python
grafico_tipo_fuego = datos_fuego.head(10)

grafico_tipo_fuego.drop('Name', axis = 1)['Att'].plot(kind='bar')
```

```python
grafico_tipo_fuego = datos_fuego.head(10)

grafico_tipo_fuego.set_index('Name', inplace=True)

grafico_tipo_fuego['Att'].plot(kind='bar')
```

Cantidad de pokemons tipo fuego por generacion

```python
conteo_tipo_fuego = datos_fuego["Generation"].value_counts()
conteo_fuego = conteo_tipo_fuego.sort_index()
plt.plot(conteo_fuego.index, conteo_fuego.values, marker='o', label="Tipo_Fuego")

plt.title("Pokemons tipo fuego por generacion")
plt.xlabel("Generacion") 
plt.ylabel("Cantidad") 

plt.grid(True)
plt.legend()

plt.show()
```
